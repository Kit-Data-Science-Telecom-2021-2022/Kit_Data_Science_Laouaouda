## Chargement des données dans un DataFrame

```python
path = './Classements'
files = [os.path.join(path, x) for x in os.listdir(path)]
stop_index = files.index(os.path.join(path, "vendeeglobe_20210127_170000.xlsx"))
files = files[:stop_index] ## only rankings before first arrival
df = pd.DataFrame()
for file in files :
    df0 = pd.read_excel(file, header=0, usecols="B:U", skiprows=4, skipfooter=6, engine="openpyxl",
                        names=['Rang', 'Nat./Voile', 'Skipper/Bateau','Heure_FR','Latitude','Longitude','Cap_30min',
                                'Vitesse_30min', 'VMG_30min','Distance_30min','Cap_last','Vitesse_last','VMG_last',
                               'Distance_last','Cap_24h','Vitesse_24h','VMG_24h','Distance_24h','DTF','DTL'])
    df0.insert(0, 'Date_classement', ''.join(re.findall('_(.*).xlsx', file)))  ## ''.joint() to convert one-element list to string
    df = df.append(df0, ignore_index=True)
```

## Nettoyage des données de classement des skippers

```python
# Supprimer les lignes avec des valeurs manquantes
df_ = df_.dropna()
df_.reset_index(drop=True, inplace=True)

# Ecrire la colonne 'Date_classement' sous format Datetime
df_['Date_classement'] = df_['Date_classement'].apply(lambda x: datetime.strptime(x, "%Y%m%d_%H%M%S"))
# ou : df_['Date_classement']=pd.to_datetime(df_['Date_classement'], format="%Y%m%d_%H%M%S")

# Tranformer toutes les valeurs de la colonne 'Rang' en numérique
df_['Rang'] = df_['Rang'].astype('int')

# Eclatement de la colonne Nat./Voile
col = df_.pop('Nat./Voile')
df_.insert(2, 'Nationalité', col.str.extract(r'\n(.*) (.*)')[0])
df_.insert(3, 'Numéro du voilier', col.str.extract(r'\n(.*) (.*)')[1].astype(int))

# Eclatement de la colonne 'Skipper/Bateau'
col = df_.pop('Skipper/Bateau')
df_.insert(2, 'Nom du skipper', col.str.extract(r'(.*)\n(.*)')[0])
df_.insert(5, 'Nom du voilier', col.str.extract(r'(.*)\n(.*)')[1])

# Nettoyage de la colonne 'Heure_FR'
df_['Heure_FR'] = df_['Heure_FR'].str.extract('^(\d*:\d*)')[0]

# Conversion du format DMS (format affiché par un GPS) au format DD (degrés décimaux)
df_['Latitude'] = df_['Latitude'].str.split('[°\']').apply(lambda x: (float(x[0]) + float(x[1])/60) * (1 if x[2] in ['E', 'N'] else -1))
df_['Longitude'] = df_['Longitude'].str.split('[°\']').apply(lambda x: (float(x[0]) + float(x[1])/60) * (1 if x[2] in ['E', 'N'] else -1))

# Nettoyage des colonnes Cap
df_[['Cap_30min', 'Cap_last', 'Cap_24h']] = df_[['Cap_30min', 'Cap_last', 'Cap_24h']].applymap(lambda x: int(x[:-1]))

# Nettoyage des colonnes Vitesses
df_['Vitesse_30min'] = df_['Vitesse_30min'].str.split().apply(lambda x: float(x[0]))
df_['Vitesse_last'] = df_['Vitesse_last'].str.split().apply(lambda x: float(x[0]))
df_['Vitesse_24h'] = df_['Vitesse_24h'].str.split().apply(lambda x: float(x[0]))

# Nettoyage des colonnes VMG
df_['VMG_30min'] = df_['VMG_30min'].str.split().apply(lambda x: float(x[0]))
df_['VMG_last'] = df_['VMG_last'].str.split().apply(lambda x: float(x[0]))
df_['VMG_24h'] = df_['VMG_24h'].str.split().apply(lambda x: float(x[0]))
                                                  
# Nettoyage des colonnes Distance
df_['Distance_30min'] = df_['Distance_30min'].str.split().apply(lambda x: float(x[0]))
df_['Distance_last'] = df_['Distance_last'].str.split().apply(lambda x: float(x[0]))
df_['Distance_24h'] = df_['Distance_24h'].str.split().apply(lambda x: float(x[0]))
                                                  
# Nettoyage de la colonne DTF
df_['DTF'] = df_['DTF'].str.split().apply(lambda x: float(x[0]))
                                                  
# Nettoyage de la colonne DTL
df_['DTL'] = df_['DTL'].str.split().apply(lambda x: float(x[0]))                                            
```

# 2) Extraction, chargement et nettoyage des données sur les caractéristiques techniques des voiliers

## Téléchargement des données sur les caractéristiques techniques des voiliers et construction du Dataset

```python
resp = requests.get("https://www.vendeeglobe.org/fr/glossaire")
b_s = BeautifulSoup(resp.content)
voiliers=[]
c=0
for item in b_s.findAll("div" , attrs={"class":"sv-u-1 sv-u-s-1-2 sv-u-m-1-3 sv-u-l-1-6"}):
    voilier = dict()
    voilier[item.find('h3').attrs['class'][0].split('__')[1]] = item.find('h3', attrs={'class':'boats-list__boat-name'}).text
    voilier[item.find('span').attrs['class'][0].split('__')[1]] = item.find('span', attrs={'class':'boats-list__skipper-name'}).text
    for i in item.findAll("li"):
        voilier[i.text.split(' : ')[0]] = i.text.split(' : ')[1]
    voiliers.append(voilier)
df1 = pd.DataFrame(voiliers)
```

```python
df1
```

```python
df1.info()
```

## Nettoyage des données sur les caractéristiques techniques des voiliers

```python
# Renommage des colonnes 'skipper-name' et 'boat-name'
df1.rename(columns={'skipper-name':'Nom du skipper', 'boat-name':'Nom du voilier'}, inplace=True)

# Nettoyage de la colonne 'Nom du skipper'
df1['Nom du skipper'] = df1['Nom du skipper'].str.title()

# Nettoyage de la colonne 'Numéro de voile'
# Extraction du numéro de voilier uniquement
col = df1.pop('Numéro de voile')
df1.insert(0, 'Numéro du voilier', col.str.extract('(\d+)$'))
# Imputation de la valeur manquante de la colonne 'Numéro de voile' à partir du DataFrame df_
df1['Numéro du voilier'].fillna(df_.loc[df_['Nom du skipper']=='Thomas Ruyant','Numéro du voilier'].iloc[0], inplace=True)
# Suppression des zéros superflues (à gauche des chiffres)
df1['Numéro du voilier'] = df1['Numéro du voilier'].astype('str').str.extract('^0*([1-9][0-9]*$)').astype(int)

# Transformation du type de données dans la colonne 'Date de lancement' en Datetime
# Le format de la méthode pd.to_datetime ne reconnait pas les noms de mois en francais. Donc, je map ces noms avec le numéro du mois correspondant avant de tranformer le type des données de la colonne 'Date de lancement' en Datetime  
dic = {'Janvier': 1,'Février': 2,'Mars': 3,'Avril':4,'Mai':5,'Juin':6,'Juillet':7,'Août':8,'Septembre':9,'Octobre':10,'Novembre':11,'Décembre':12}
df1['Date de lancement'] = df1['Date de lancement'].str.split().apply(lambda x: x[0]+'/'+str(dic[x[1]])+'/'+x[2])
df1['Date de lancement'] = pd.to_datetime(df1['Date de lancement'], format="%d/%m/%Y")

# Nettoyage des colonnes 'Longueur', 'Largeur'
df1['Longueur'] = df1['Longueur'].str.replace(',','.').str[:-2].astype(float)
df1['Largeur'] = df1['Largeur'].str.replace(',','.').str[:-2].astype(float)
df1["Tirant d'eau"] = df1["Tirant d'eau"].str.replace(',','.').str[:-2].astype(float)

# Nettoyage de la colonne 'Déplacement (poids)'
df1['Déplacement (poids)'].replace(['NC', 'nc'],np.nan, inplace=True)
df1['Déplacement (poids)'] = df1['Déplacement (poids)'].str.replace(',','.').str.extract('^([\d\.]*)').astype(float)

# Nettoyage de la colonne 'Hauteur mât'
df1['Hauteur mât'] = df1['Hauteur mât'].str.extract('^(\d*)').astype(int)

# Nettoyage de la colonne 'Voile quille'
df1['Voile quille'] = df1['Voile quille'].str.lower()

# Nettoyage des colonnes 'Surface de voiles au près', 'Surface de voiles au portant'
df1["Surface de voiles au près"] = df1["Surface de voiles au près"].str[:-2].astype(int)
df1["Surface de voiles au portant"] = df1["Surface de voiles au portant"].str[:-2].astype(int)

# Recodage binaire de la colonne 'Nombre de dérives'
df1.rename({'Nombre de dérives':'Foils'}, axis=1, inplace=True)
df1['Foils'] = df1['Foils'].apply(lambda x: 'NON' if x!='foils' else 'OUI')
```

## Corrélation et régression linéaire entre le classement (rang) et les différentes vitesses utiles (VMG) des voiliers

### Corrélation le rang et les différentes vitesses utiles (VMG) des voiliers

Le coefficient de corrélation linéaire de Pearson est une estimation (corrélation empirique) de la corrélation linéaire entre 2 variables numériques.  
Un test de Student permet de vérifier si la corrélation est significative ou pas (il existe réellement une corrélation entre les 2 variables au niveau de la population).  
**NB** : Le coefficient de corrélation de Pearson permet d'évaluer uniquement la corrélation de type *linéaire* entre 2 variables. Par exemple, les variables X et X^2 sont corrélés quadratiquement, or, le coefficient de corrélation de Pearson (peut être vu comme le cosinus de l'angle entre les 2 vecteurs centrés) est nul.

```python
pd.DataFrame({col:pearsonr(df_[col], df_['Rang']) for col in ('VMG_30min', 'VMG_last', 'VMG_24h')}, 
             index=['coefficient de corrélation de Pearson avec la variable Rang', 'p-value associée au t-test'])
```

La variable 'Rang' est faiblement corrélée avec les variables 'VMG_30min' (-0.27), 'VMG_last' (-0.30) et 'VMG_24h' (-0.36).  
La corrélation entre la variable 'Rang' et les 3 variables est à chaque fois significative.

### Régression linéaire sur tous les points entre le rang et les différentes vitesses utiles (VMG) des voiliers

```python
# Nuage de points
fig, ax = plt.subplots(1,3, figsize=(15,5))
for i, col in enumerate(['VMG_30min', 'VMG_last', 'VMG_24h']):
    ax[i].scatter(df_[col], df_['Rang'])
    ax[i].set_xlabel(col)
    ax[i].set_ylabel('Rang')
    ax[i].set_title(f'Nuage de points {col}/Rang ');
```

Dans la régression linéaire simple (avec une seule covariable), le coefficient de détermination R^2 est égal au carré du coefficient de corrélation de Pearson r entre la variable indépendante X et la variable dépendante Y.  
Le test de Fisher (test de significativité globale des coefficients de régression) permet de tester si notre régression est globalement significative ou pas. L'hypothèse nulle est la nullité de tous les coefficients de régression.  
Le test de Student (test de significativité individuelle des coefficients de régression) permet de tester si un coefficient de régression est significatif ou pas. L'hypothèse nulle est la nullité de ce coefficient de régression.  
Le test de Jarque-Bera vérifie l'adéquation (conformité) d'une distribution empirique à la distribution gaussienne. L'hypothèse nulle est l'adéquation des données à une loi normale.
Les moments d'ordre 3 et 4, respectivement Skewness (coefficient d'asymetrie) et Kurtosis (coefficient d'aplatissement) donnent une idée sur la forme de la courbe de la distribution des erreurs.

```python
# add intercept to the model
X = sm.add_constant(df_['VMG_30min'])
# fit model
y = df_['Rang']
reg = sm.OLS(y, X)
reg_fit = reg.fit()
print(reg_fit.summary())
```

Le coefficient d'adéquation est petit (0.076). C'est à dire, que uniquement 7.6% de la variabilité de la variable réponse Y est expliquée par le modèle. Donc, notre modèle a un pouvoir prédictif faible.  
La p-value du test de Fisher est très inférieure à 5%. On rejette donc l'hypothèse nulle du test. Ainsi, notre régression est globalement significative. Egalement, la corrélation entre la variable cible et la covariance est significative.  
La p-value du test de Student associée à l'intercept est très inférieure à 5%. Par conséquent, l'intercept du modèle linéaire est significatif.  
La p-value du test de Student associée à la pente est très inférieure à 5%. Par conséquent, la pente du modèle linéaire est significative.  
La p-value du test de Jarque-Bera est trop inférieure à 5%. On rejette donc l'hypothèse nulle selon laquelle les erreurs de la régresssion linéaire suivent une loi normale.

### Régression sur les données agrégées (par nom de voilier) entre le rang et les différentes vitesses utiles (VMG) des voiliers

Une autre approche serait d'effectuer la régression avec les mêmes variable mais cette fois sur des données regroupées par skipper/voilier. Pour cela, on calcule le rang moyen et la VMG moyenne pour chaque voilier.  
**NB** : Il est plus judicieux dans notre cas d'utiliser la moyenne arithmétique pour calculer la VMG moyenne et pas la moyenne harmonique

```python
#Construnction du dataframe qui donne le rang moyen et la vitesse moyenne par Voilier
df_agg=df_[['Rang','VMG_30min', 'VMG_last', 'VMG_24h', 'Numéro du voilier']] \
        .groupby(by='Numéro du voilier').agg({'Rang':'mean','VMG_30min':'mean', 'VMG_last':'mean', 'VMG_24h':'mean'})
df_agg.head()
```

**NB** : Il aurait été plus judicieux de supprimer les données des skippers qui n'ont pas fini la course. Or, en faisant cela, on aura un nombre d'observations inférieur à 30 et donc notre modèle de régression fournira des résultats biaisées.

```python
# Nuage de points
fig, ax = plt.subplots(1,3, figsize=(15,5))
for i, col in enumerate(['VMG_30min', 'VMG_last', 'VMG_24h']):
    lin_reg = LinearRegression()
    lin_reg.fit(df_agg[col].to_frame(), df_agg['Rang'])
    y_pred = lin_reg.predict(df_agg[col].to_frame())
    ax[i].scatter(df_agg[col], df_agg['Rang'])
    ax[i].plot(df_agg[col], y_pred, 'b', label='Regression line')
    ax[i].set_xlabel(col)
    ax[i].set_ylabel('Rang')
    ax[i].set_title(f'Nuage de points {col}/Rang ')
    ax[i].legend(loc='lower left')
plt.show()
```

```python
for i, col in enumerate(['VMG_30min', 'VMG_last', 'VMG_24h']):
    print(f"Le coefficient de détermination de la régression de la variable Rang sur la variable {col} est : {lin_reg.score(df_agg[col].to_frame(), df_agg['Rang']):.2%}")
```

On obtient un bon coefficient d'adéquation pour les 3 régressions.

## KMeans sur les caractéristiques techniques des voiliers

L’algorithme des K-moyennes (K-means) est un algorithme d'apprentissage non supervisé ayant pour objectif de séparer les données en groupes homogènes ayant des caractéristiques communes.

```python
# Extraction de certaines caractéristiques des voiliers
df1_ = df1[['Longueur', 'Largeur', "Tirant d'eau", 'Déplacement (poids)', 'Foils', 'Hauteur mât', 'Voile quille', \
            'Surface de voiles au près', 'Surface de voiles au portant']]
df1_.dropna(inplace=True)
df1_.reset_index(drop=True, inplace=True)
```

### Prétraitement des données

Pour appliquer la méthode Kmeans, il faut que les variables soient numérique. Or, on a 2 variables qualitatives, à savoir 'Foils' et 'Voile quille'.  
La variable 'Foils' a deux modalités. On peut donc faire un recodage numérique binaire.  
La variable 'Voile quille' a plusieurs modalités sans ordre naturel (variable qualitative nominale). Recoder numériquement ces modalités n'est pas la meilleure solution (car le recodage numérique sous-entend un classement des modalités ce qui va à l'encontre de la nature de la variable de type nominale). La meilleure solution serait donc l'encodage à chaud (one-hot-encoding). Cette variable sera éclatée en variables binaires (dummy variables) dont le nombre est égal au nombre des modalités de la variable de départ. Or, les catégories de cette variable sont nombreuses et non équilibrées. Par conséquent, on propose de regrouper les modalités faiblement représentées dans la classe 'Autre matériel'.  
**NB** : Dans le cadre du one-hot-encoding, la création de variables indicatrices dont le nombre est égal au nombre des modalités de la variable de départ crée un problème de multicolinéarité (la valeur d'une variable binaire peut toujours être déduite des n-1 valeurs de variables binaires restanes) qui impacte la méthode KMeans. Or, ceci ne présentera pas un problème dans notre cas puisqu'on appliquera par la suite l'ACP qui élimine la multicolinéarité.

```python
# Recodage de la variable 'Foils' en binaire
df1_['Foils'] = df1_['Foils'].apply(lambda x: 1 if x=='Oui' else 0)
df1_

# Création de nouvelles classes à partir des anciennes catégories pour la variable 'Voile quille'
df1_['Voile quille'] = df1_['Voile quille'].replace(list(df1_['Voile quille'].value_counts()[2:].index), 'autre matériel')
# One-hot-encoding
lb = LabelBinarizer()
df0 = pd.DataFrame(lb.fit_transform(df1_['Voile quille'].to_numpy()), columns=lb.classes_)
df0 = df0[['acier forgé', 'acier', 'autre matériel']]
df1_.drop('Voile quille', axis=1, inplace=True)
df1_ = df1_.join(df0, how='right')
```

On élimine l'effet d'échelle en standardisant nos variables

```python
scaler = StandardScaler()
scaled = scaler.fit(df1_) # Compute the mean and std to be used for later scaling
std_df1_ = pd.DataFrame(scaled.transform(df1_), index=df1_.index, columns=df1_.columns)
```

```python
std_df1_.head()
```

### Détermination du nombre de voisins optimal et application de la méthode KMeans

On souhaite déterminer l'hyperparamètre k (nombre de clusters) optimal pour la méthode KMeans. Pour cela, on trace pour plusieurs valeurs de k, le score du clustering. On choisira comme mètrique pour évaluer le score : la distortion totale. Pour un cluster $C$ de centroide $c$ entouré par les points $(x^{c}_i)_{i=0}$, la distortion $d(c)$ se définit comme :
$$
    d(c) = \sum_i |x_i - c|^2,
$$
En utilisant 'la méthode du coude', on détermine k optimal.

```python
def distortion(data, clustering):
    X_new = clustering.fit_transform(data)
    distortion = clustering.inertia_
    return distortion

min_k = 1
max_k = 10
elbow_scores = [distortion(std_df1_, KMeans(n_clusters=k, random_state=1)) for k in list(range(min_k, max_k))]
print(elbow_scores)
plt.plot(list(range(min_k, max_k)), elbow_scores)
plt.xlabel("n_clusters")
plt.ylabel("Distortion");
```

On remarque une cassure dans la courbe de distortion pour la valeur k=5. La méthode du coude nous dit alors que le nombre de clusters optimal est 5. 

```python
kmeans_algo = KMeans(n_clusters=5, init='k-means++', random_state=1, verbose=1) # Compute k-means clustering (fit estimator)
X_new = kmeans_algo.fit_transform(df1_) # transform X to cluster-distance space
print(f"\n\nLe nombre d'itérations : {kmeans_algo.n_iter_}", end='\n\n')
print(f"Les labels (numéro du cluster) des observations : {kmeans_algo.labels_}", end='\n\n\n')
print(f"Les centroïdes des clusters : {kmeans_algo.cluster_centers_}", end='\n\n\n')
print(f"L'inertie (variance totale) est égale à : {kmeans_algo.inertia_}", end='\n\n')  ## inertia(SSE)
```

### Réalisation d'une ACP pour visualiser les résultats

En grande dimension (11 variables dans notre cas), il n'est pas possible de visualiser les résultats de KMeans.  
Une solution serait de faire une réduction de dimension en appliquant l'ACP afin de créer 2 variables synthétiques (axes factoriels).  
**NB** : On a appliqué l'ACP uniquement dans un but de visualisation.

```python
pca = PCA(n_components=2) #2-dimensional PCA
df_pca = pd.DataFrame(pca.fit_transform(df1_), index=std_df1_.index, columns=['PC1', 'PC2'])
plt.scatter(df_pca.loc[kmeans_algo.labels_==0, 'PC1'], df_pca.loc[kmeans_algo.labels_==0, 'PC2'], label='Class 1', c='red')
plt.scatter(df_pca.loc[kmeans_algo.labels_==1, 'PC1'], df_pca.loc[kmeans_algo.labels_==1, 'PC2'], label='Class 2', c='green')
plt.scatter(df_pca.loc[kmeans_algo.labels_==2, 'PC1'], df_pca.loc[kmeans_algo.labels_==2, 'PC2'], label='Class 3', c='blue')
plt.scatter(df_pca.loc[kmeans_algo.labels_==3, 'PC1'], df_pca.loc[kmeans_algo.labels_==3, 'PC2'], label='Class 4', c='orange')
plt.scatter(df_pca.loc[kmeans_algo.labels_==4, 'PC1'], df_pca.loc[kmeans_algo.labels_==4, 'PC2'], label='Class 5', c='cyan')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='upper right');
```

## Impact de la présence d'un foil sur le classement et sur les différentes vitesses utiles des voiliers

```python
# Construction d'un Dataframe composé des colonnes 'Foils' et 'Rang final'
df_tmp = df_fus.loc[df_fus['Date_classement']=='2021-01-27 14:00:00',['Numéro du voilier','Foils','Rang']].rename(columns={'Rang':'Rang final'})
df_tmp.set_index('Numéro du voilier', inplace=True)

# Construction d'un Dataframe composé des colonnes 'VMG_last_mean' et 'VMG_last_max'
df_agg = df_[['Rang', 'VMG_last', 'Numéro du voilier']] \
        .groupby(by='Numéro du voilier').agg({'Rang':'mean','VMG_last':['mean','max']})
df_agg.columns = ['Rang', 'VMG_last_mean', 'VMG_last_max']

# Fusion des 2 DataFrames
df_f = pd.merge(df_tmp, df_agg, left_on='Numéro du voilier', right_on='Numéro du voilier')
df_f.drop('Rang', axis=1, inplace=True)
df_f.head()
```

&emsp; Une boite-à-moustaches donne des informations graphiques sur la position (à travers la médiane) ,la dispersion (à travers l'écart interquartile et la distance entre les moustaches), l'asymétrie (en comparant la médiane à la moyenne) et les points aberrants (points représentés au-delà des moustaches).  
&emsp; Dans notre cas, on représente le boxplot de la variable 'Rang final' selon les 2 modalités de la variable 'Foils' pour voir s'il y une relation significative entre les 2 variables.  

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df_f['Rang final'], y=df_f['Foils'])
plt.axvline(x=df_f.loc[df_f['Foils']=='OUI','Rang final'].mean(), color='red', linestyle='--', label='Foils_mean')
plt.axvline(x=df_f.loc[df_f['Foils']=='NON','Rang final'].mean(), color='cyan', linestyle='--', label='No_Foils_mean')
plt.legend(loc='upper right');
```

On remarque que les 2 plots n'ont pas d'outliers et à peu près la même dispersion. La distribution du 2ème groupe (No_Foils) a une assymètrie négative (moyenne < médiane) et la distribution du 1er groupe est symètrique.  
La médiane du 1er groupe (Foils) est inférieure au quartile inférieure du 2ème groupe (No_Foils). Par conséquent, on peut conclure que la présence d'un foils influence significativement le rang final des skippers. En effet, les voiliers équipés d'un foils ont tendance à avoir de meilleurs classements.

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
sns.boxplot(x=df_f['VMG_last_mean'], y=df_f['Foils'])
plt.axvline(x=df_f.loc[df_f['Foils']=='OUI','VMG_last_mean'].mean(), color='red', linestyle='--', label='Foils_mean')
plt.axvline(x=df_f.loc[df_f['Foils']=='NON','VMG_last_mean'].mean(), color='cyan', linestyle='--', label='No_Foils_mean')
plt.legend(loc='upper left')
plt.subplot(1,2,2)
sns.boxplot(x=df_f['VMG_last_max'], y=df_f['Foils'])
plt.axvline(x=df_f.loc[df_f['Foils']=='OUI','VMG_last_max'].mean(), color='red', linestyle='--', label='Foils_mean')
plt.axvline(x=df_f.loc[df_f['Foils']=='NON','VMG_last_max'].mean(), color='cyan', linestyle='--', label='No_Foils_mean')
plt.legend(loc='upper left');
```

La médiane du 2ème groupe (No_Foils) est inférieure au quartile inférieure du 1er groupe (Foils). Par conséquent, on peut conclure que la présence d'un foils influence significativement la VMG (calculée depuis le dernier classement) moyenne des voiliers. La médiane du 2ème groupe (No_Foils) est inférieure à la moustache inférieure (minimum de la série) du 1er groupe (Foils). Par conséquent, on peut conclure que la présence d'un foils inmpacte significativement la VMG (calculée depuis le dernier classement) maximale des voiliers. En effet, les voiliers équipés d'un foils peuvent atteindre une VMG maximale significativement plus grande que celle atteinte par les voiliers sans foils.
